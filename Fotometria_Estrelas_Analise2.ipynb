### 4. Pré-processamento dos Dados

Neste item, a coluna `MAG_APER_COR_3_0` é dividida em 57 colunas separadas, cada uma contendo um valor de fotometria. Isso é feito usando o método `str.split(expand=True)`, que divide a string em múltiplas colunas com base nos espaços. Após a divisão, os valores são convertidos para numéricos usando `pd.to_numeric`.

O DataFrame resultante tem 63 colunas, incluindo as novas colunas de fotometria (`Fotometria_1`, `Fotometria_2`, etc.). Esse pré-processamento é crucial para análises posteriores, pois permite que cada valor de fotometria seja tratado individualmente.

```python
#A coluna `MAG_APER_COR_3_0` contém múltiplos valores separados por espaços. Vamos dividi-la em 57 colunas separadas.
#Criamos uma lista de nomes para as novas colunas.

colunas_fotometria = [f'Fotometria_{i+1}' for i in range(57)]
```

```python
# Verificar as colunas do DataFrame
print("Colunas do DataFrame:")
print(df.columns)
```

```python
#Dividimos a coluna `MAG_APER_COR_3_0` em 57 colunas.
df[colunas_fotometria] = df['MAG_APER_COR_3_0'].str.split(expand=True)
```

```python
#Convertemos os valores para numéricos.
df[colunas_fotometria] = df[colunas_fotometria].apply(pd.to_numeric)
```

```python
#Exibimos as primeiras linhas após o pré-processamento.

print("\nDataFrame após divisão da coluna de fotometria:")
df.head()
```

### 5. Visualização das 10 Primeiras Fotometrias

O objetivo deste item é visualizar as magnitudes fotométricas das 10 primeiras estrelas em todas as 57 bandas fotométricas. A visualização é feita por meio de gráficos de linhas, onde cada linha representa uma estrela e o eixo X representa as bandas fotométricas (pelo seus respectivos nomes).

```python
#Vamos plotar as 10 primeiras fotometrias como gráficos de linhas.
plt.figure(figsize=(12, 8))
```

```python
# Carregar o arquivo CSV
df_filters = pd.read_csv('/content/804024.csv')

# Extrair os nomes das bandas fotométricas
filter_names = df_filters['name'].tolist()

# Supondo que df seja o DataFrame que contém as magnitudes das estrelas
# e que as colunas de fotometria sejam nomeadas como 'Fotometria_1', 'Fotometria_2', ..., 'Fotometria_57'
columns_fotometria = [f'Fotometria_{i+1}' for i in range(57)]

#Loop para plotar as 10 primeiras estrelas.

# Criar a figura
plt.figure(figsize=(15, 8))

# Plotar as 10 primeiras estrelas
for i in range(10):
    plt.plot(filter_names, df.loc[i, columns_fotometria], label=f'Estrela {i+1}')

# Configurações do eixo X
plt.xlabel('Banda Fotométrica (Nome)')
plt.ylabel('Magnitude')
plt.title('Fotometria das 10 Primeiras Estrelas')
plt.xticks(rotation=90)  # Rotaciona os rótulos do eixo X para melhor visualização

# Invertendo o eixo Y para valores decrescentes
plt.gca().invert_yaxis()

# Adicionando a legenda e ajustando o layout
plt.legend()
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

#### 8.2. Normalização dos Dados

Para comparar padrões de forma independente do brilho absoluto, aplica-se normalização **Min-Max** (escala [0, 1]), que preserva a forma das curvas fotométricas enquanto elimina diferenças de magnitude global.

**Processo de Normalização:**


*  ** Seleção das features:** Apenas as colunas de magnitudes (Fotometria_1 a Fotometria_57) são normalizadas.
*   **Transformação:**
  *   **Mínimo = 0** (valor mais brilhante em uma estrela).
  *   **Máximo = 1** (valor mais fraco na mesma estrela).
  *   Cada magnitude é recalculada como:

  Mag_normalizada = Mag_original - min(Mag_estrela) / max(Mag_estrela) - min(Mag_estrela)

*   **Novo DataFrame:** Os dados normalizados são armazenados em um novo DataFrame (df_normalizado), mantendo a estrutura original.

**Visualização Comparativa:**

* Subplot superior: Dados originais (mesmo que o código inicial).
* Subplot inferior: Dados normalizados.
  * Efeito da normalização: Todas as curvas são "achatadas" para o intervalo [0, 1], destacando variações relativas dentro de cada estrela.
  * Padrões destacados: Formas de distribuição espectral (p.ex., picos em bandas específicas) tornam-se mais visíveis, pois a normalização remove o viés de brilho total.

**Vantagens da Normalização:**



1.   Comparabilidade: Permite comparar estrelas de brilhos intrínsecos diferentes em igualdade de condições.
2.   Identificação de assinaturas espectrais: Padrões como excesso de infravermelho (poeira) ou absorção (linhas moleculares) podem ser mais facilmente detectados.
3.   Pré-processamento para ML: Algoritmos de classificação ou clustering dependem de dados em escalas similares para funcionar adequadamente.








              
                    
                  






### 9. Redução de Dimensionalidade com PCA

Este código aplica Análise de Componentes Principais (PCA) para identificar as bandas fotométricas mais discriminativas, reduzindo a dimensionalidade dos dados normalizados.

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Carregar dados normalizados (do código anterior)
X_normalized = df_normalizado[columns_fotometria].values

# Aplicar PCA
pca = PCA(n_components=2)  # Reduzir para 2 componentes para visualização
X_pca = pca.fit_transform(X_normalized)

# Visualizar a variância explicada por cada componente
plt.figure(figsize=(10, 5))
plt.bar(range(2), pca.explained_variance_ratio_, alpha=0.7, label='Variância explicada')
plt.xlabel('Componente Principal')
plt.ylabel('Proporção da Variância Explicada')
plt.title('Variância Explicada pelo PCA')
plt.legend()
plt.grid(True)
plt.show()

# Plotar as estrelas no espaço das 2 primeiras componentes
plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.7, edgecolors='k')
plt.xlabel('Primeira Componente Principal (PC1)')
plt.ylabel('Segunda Componente Principal (PC2)')
plt.title('Projeção das Estrelas em 2D (PCA)')
plt.grid(True)
plt.show()

# Identificar bandas mais importantes em cada componente
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
important_bands_pc1 = np.abs(loadings[:, 0]).argsort()[-5:][::-1]  # Top 5 em PC1
important_bands_pc2 = np.abs(loadings[:, 1]).argsort()[-5:][::-1]  # Top 5 em PC2

print("\nBandas mais importantes na PC1:")
print([filter_names[i] for i in important_bands_pc1])

print("\nBandas mais importantes na PC2:")
print([filter_names[i] for i in important_bands_pc2])
```

* PCA reduz os dados a 2 componentes principais, mantendo a maior variância possível.
* O gráfico de variância ajuda a entender quanta informação cada componente retém.
* O scatter plot mostra como as estrelas se distribuem no novo espaço.
* As bandas mais influentes em cada componente são identificadas, indicando quais filtros fotométricos melhor discriminam as estrelas.
